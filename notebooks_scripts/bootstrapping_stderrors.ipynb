```python
def independent_bootstrapping(results, metric, tries=1000):
    metrics = []
    for i in range(tries):
        idxs = np.random.choice(results.shape[0], (results.shape[0]))
        new_results = results[idxs]
        metrics.append(metric(new_results))
    metrics = np.asarray(metrics)
    return np.mean(metrics), np.std(metrics)

```

```python
def top1_median(x):
    return np.median(x[:, 0])

def top1_below2(x):
    return (100 * (x[:, 0] < 2).sum() / len(x[:, 0]))

def top5_median(x):
    return np.median(np.min(x[:, :5], axis=1))

def top5_below2(x):
    x5 = np.min(x[:, :5], axis=1)
    return (100 * (x5 < 2).sum() / len(x5))
```

```python
def read_strings_from_txt(path):
    # every line will be one element of the returned list
    with open(path) as file:
        lines = file.readlines()
        return [line.rstrip() for line in lines]

names_no_rec_overlap = read_strings_from_txt(f'data/splits/timesplit_test_no_rec_overlap')
```

```python
for method, rmsds in all_rmsds_list:
    performance_metrics = {}
    
    keep = np.asarray([t in names_no_rec_overlap for t in names[method]])
    rmsds = rmsds[keep]
    
    top1_rmsds = rmsds[:,0]
    performance_metrics.update({
        'median_rmsd': (np.median(top1_rmsds), independent_bootstrapping(rmsds, top1_median)),
        'rmsds_below_2': (100 * (top1_rmsds < 2).sum() / len(top1_rmsds), independent_bootstrapping(rmsds, top1_below2)),
        })
    top5_rmsds = np.min(rmsds[:, :5], axis=1)
    performance_metrics.update({
        f'top5_median_rmsd': (np.median(top5_rmsds), independent_bootstrapping(rmsds, top5_median)),
        f'top5_rmsds_below_2': ((100 * (top5_rmsds < 2).sum() / len(top5_rmsds)).__round__(2), independent_bootstrapping(rmsds, top5_below2)),
    })
    print('\n')
    print(method)
    for k in performance_metrics:
        print(k, performance_metrics[k])
```
